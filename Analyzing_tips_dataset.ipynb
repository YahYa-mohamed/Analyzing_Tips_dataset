#  Seaborn library
# Documantation:
> - Total bill : (cost of the meal), including tax, in US dollars.
- tip : tip (gratuity) in US dollars.
- sex : sex of person paying for the meal (0=male, 1=female).
- smoker : smoker in party? (0=No, 1=Yes).
- day : 3=Thur, 4=Fri, 5=Sat, 6=Sun.
- time : 0=Day, 1=Night.
- size : size of the party.

# The dataset 
the dataset is `tips`
> The dataset from [here](https://www.kaggle.com/ranjeetjain3/seaborn-tips-dataset)

```python
# import seaborn 
import seaborn as sns
# use %matplotlib inline to display the graphs on jupyter notebook
%matplotlib inline
#use set() function to set the details about graphs show
sns.set(rc={'figure.figsize':[8,8]},font_scale=1.5)
```

**Loading dataset from library**

```python
df=sns.load_dataset('tips')
```

```python
#display first five rows in the data 
df.head()
```

```python
#display the last five rows in the data
df.tail()
```

```python
#do some statistics on data
df.describe()
```

```python
# show some information about data
df.info()
```

```python
# change the type of total_bill from float to integer
df.total_bill=df.total_bill.astype(int)
```

```python
#check for changing
df.total_bill
```

# Types of Distribution plots

>**Univariate plots:**
 - distplot
 - kdeplot<br>
**Bivariate plots:**
 - jointplot
 - pairplot 

**1-Univariate plots(distplot)**

# what is the highest total_bill?

```python
sns.distplot(df.total_bill,color='g',bins=30,kde=False);
```

# what is the highest size?

```python
sns.distplot(df['size'],color='y',kde=False,bins=35);
```

# what is the highest tip?

```python
sns.distplot(df['tip'],color='r',kde=False,bins=40);
```

**2-Univariate plots(kdeplot)**

```python
sns.kdeplot(df.total_bill,color='g');
```

```python
sns.kdeplot(df.tip,color='r',shade=True);
```

```python
sns.kdeplot(df['size'],color='k',shade=True);
```

**1-Bivariate(jointplot)**

```python
# relationship between total_bill and tip
sns.jointplot(x='total_bill',y='tip',data=df,color='r',kind='scatter',height=10);
```

```python
#relationship between total_bill and tip using reg kind
sns.jointplot(x='total_bill',y='tip',data=df,color='k',height=10,kind='reg');
#also can we use any kind of it ('scatter', 'reg', 'resid', 'kde', 'hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,color='y',kind='resid',height=10);
```

```python
#using kde kind that stands for kernal density estmation
sns.jointplot(x='total_bill',y='tip',data=df,color='g',kind='kde',height=10);
```

```python
# using hex kind
sns.jointplot(x='total_bill',y='tip',color='b',height=10,data=df,kind='hex');
```

```python
cmap=sns.cubehelix_palette(as_cmap=True,light=1,dark=2)
sns.jointplot(x='total_bill',y='tip',data=df,color='y',kind='reg',height=10)
sns.jointplot(x='total_bill',y='tip',data=df,color='r',kind='kde',height=10)
```

```python
cmap=sns.cubehelix_palette(as_cmap=True,light=1,dark=0,reverse=True)
sns.jointplot(x='total_bill',y='tip',data=df,cmap=cmap,kind='hex',height=10)
sns.jointplot(x='total_bill',y='tip',data=df,cmap=cmap,kind='kde',height=10)
```

**2-Bivariate(pairplot)**

```python
sns.pairplot(df);
```

```python
sns.pairplot(df,height=3.7,aspect=1.2);
```

```python
sns.pairplot(df,vars=['total_bill','tip'],height=3.5,aspect=1.5);
```

```python
sns.pairplot(df,hue='sex',height=3,aspect=1.2);
```

```python
sns.pairplot(df,hue='day',height=3,aspect=1,diag_kind='hist'); 
```

```python
sns.pairplot(df,hue='time',height=3,aspect=1.4,diag_kind='hist');
```

```python
sns.pairplot(df,hue='smoker',aspect=1.2,height=3,diag_kind='hist');
```

```python
sns.pairplot(df,hue='smoker',aspect=1.2,height=3);
```

```python
sns.pairplot(df,hue='size',aspect=1.2,height=3,diag_kind='hist');
```

```python
g=sns.PairGrid(df,height=3,aspect=1.4)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot)
```

```python
g=sns.PairGrid(df,height=3,aspect=1.5)
g.map_diag(sns.kdeplot,shade=True)
g.map_offdiag(sns.kdeplot,shade=True)
```

```python
cmap=sns.cubehelix_palette(n_colors=10,dark=0,light=1,as_cmap=True,reverse=True)
g=sns.PairGrid(df,height=3,aspect=1.5)
g.map_diag(sns.kdeplot,shade=True,color='r')
g.map_offdiag(sns.kdeplot,shade=True,cmap=cmap)
```
